```python
bin_masa=60
histtt,edgesm=histogram(log10(tabla[:,0]), bins=bin_masa, normed=True)
hist_vel=np.zeros([bin_masa])
hist_pos=np.zeros([bin_masa])
hist_masa=np.zeros([bin_masa])
error_vel=np.zeros([bin_masa])
error_pos=np.zeros([bin_masa])
error_masa=np.zeros([bin_masa])
for i in range(bin_masa):
    indexs=where((log10(tabla[:,0])>edgesm[i]) & (log10(tabla[:,0])<edgesm[i+1]))[0]
    datosv=tabla[indexs,2]
    datosp=tabla[indexs,1]
    datosm=tabla[indexs,3]
    hist_vel[i]=revisar(mean(datosv))
    hist_masa[i]=revisar(mean(datosm))
    hist_pos[i]=revisar(mean(datosp))
    error_vel[i]=revisar(std(datosv))
    error_pos[i]=revisar(std(datosp))
    error_masa[i]=revisar(std(datosm))
```

```python
def revisar(a):
    if(isnan(a)):
        return 0
    else:
        return a
```

```python
fig1=figure()
filename="logm-deltav"
ax1=axes()
ax1.set_title("$\mathrm{Comparacion\ }v\ \Omega_0+5\%$",fontsize=20)
ax1.set_xlabel("$\log_{10}(M/M_{\odot})$",fontsize=20)
ax1.set_ylabel("$\Delta |v| [km/s]$",fontsize=15)
errorbar(edgesm[:bin_masa],hist_vel,yerr=error_vel,color='g')
savefig(filename + '-mas'+ '.png',format = 'png', transparent=True)
```
